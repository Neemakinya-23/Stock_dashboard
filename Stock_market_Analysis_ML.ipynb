```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
```

```python
# Define stock ticker, start, and end dates
ticker = "AAPL"  # Apple Inc.
start_date = "2000-01-01"
end_date = "2025-01-01"

# Fetch historical data
stock_data = yf.download(ticker, start=start_date, end=end_date)

# Optional: Rename other columns for clarity
stock_data.columns = ["Close", "High", "Low", "Open", "Volume"]

# Ensure the dataframe is sorted by date
stock_data = stock_data.sort_index()

# Display last few rows
print(stock_data.tail())
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_average = stock_data.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    stock_data[ratio_column] = stock_data["Close"]/rolling_average["Close"]

    trend_column = f"Trend_{horizon}"
    stock_data[trend_column] = stock_data.shift(1).rolling(horizon).sum()["Target"]

    new_predictors +=[ratio_column,trend_column]
```

```python
stock_data.head()
```

```python
stock_data = stock_data.dropna()
stock_data.head()
```

```python
model = RandomForestClassifier(n_estimators = 200,min_samples_split = 50 ,random_state = 1)
```

```python
def predict(train,test,predictors,model2):
    model.fit(train[predictors],train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index = test.index,name ="Predictions")
    combined = pd.concat([test["Target"],preds],axis = 1)
    return combined
```

```python
Predictions = backtest(stock_data,model,new_predictors)
```

```python
# Compute MACD (12-day EMA - 26-day EMA) and Signal Line (9-day EMA of MACD)
stock_data["EMA_12"] = stock_data["Close"].ewm(span=12, adjust=False).mean()
stock_data["EMA_26"] = stock_data["Close"].ewm(span=26, adjust=False).mean()
stock_data["MACD"] = stock_data["EMA_12"] - stock_data["EMA_26"]
stock_data["Signal_Line"] = stock_data["MACD"].ewm(span=9, adjust=False).mean()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, stock_data["MACD"], label="MACD", color="blue")
plt.plot(stock_data.index, stock_data["Signal_Line"], label="Signal Line", color="red", linestyle="dashed")
plt.axhline(0, color="black", linestyle="dotted")
plt.title("MACD Indicator")
plt.xlabel("Date")
plt.ylabel("MACD Value")
plt.legend()
plt.grid()
plt.show()
```
